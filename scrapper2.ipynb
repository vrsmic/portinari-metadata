```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Configuração básica do driver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    url = "https://www.docvirt.com/docreader.net/DocReader.aspx?bib=coportinari&id=4645702686696&pagfis=1"
    driver.get(url)

    # Aguarda o carregamento do AJAX (o site é lento devido ao framework Telerik)
    time.sleep(0.9)

    # Utilizando o find_element como você solicitou
    # O primeiro argumento é a estratégia (By.ID), o segundo é o nome do ID
    textarea_element = driver.find_element(By.ID, "TextoDigitadoTxt")
    conteudo = textarea_element.get_attribute("value")

    pasta_element = driver.find_element(By.ID, "PastaTxt")
    texto_pasta = pasta_element.text  # Ou pasta_element.get_attribute("innerText")


    if conteudo:
        print("Texto encontrado:")
        print(conteudo)
        print("Texto da pasta:")
        print(texto_pasta)

    else:
        print("O elemento foi encontrado, mas ainda está vazio. Tente aumentar o time.sleep.")

finally:
    driver.quit()
```

```python
def extract_text_from_page(url):
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        driver.get(url)
        time.sleep(1.0)
        textarea_element = driver.find_element(By.ID, "TextoDigitadoTxt")
        conteudo = textarea_element.get_attribute("value")

        pasta_element = driver.find_element(By.ID, "PastaTxt")
        texto_pasta = pasta_element.text  

        linha_df = {
            "titulo": texto_pasta,
            "conteudo": conteudo
        }

        return linha_df
    
    except Exception as e:
        print(f"Erro ao extrair dados da página")
        return {"titulo": None, "conteudo": None}

    finally:
        driver.quit()
```

```python
import pandas as pd
df_dados = pd.DataFrame(columns=["titulo", "conteudo"])
```

```python
for i in range(1, 50):
    url = f"https://www.docvirt.com/docreader.net/DocReader.aspx?bib=coportinari&id=4645702686696&pagfis={i}"
    linha_df = extract_text_from_page(url)
    df_dados = pd.concat([df_dados, pd.DataFrame([linha_df])], ignore_index=True)
    
```

```python
import asyncio
import random
import pandas as pd
from playwright.async_api import async_playwright, TimeoutError as PlaywrightTimeout
from tqdm.notebook import tqdm

BASE_URL = "https://www.docvirt.com/docreader.net/DocReader.aspx"
BIB = "coportinari"
DOC_ID = "4645702686696"
TOTAL_PAGINAS = 50
PARALLEL_PAGES = 3 

async def extract_page(context, pagina, max_retries=5): # Aumentei retries para 5
    url = f"{BASE_URL}?bib={BIB}&id={DOC_ID}&pagfis={pagina}"
    
    for tentativa in range(max_retries):
        page = await context.new_page()
        try:
            # Abortar recursos desnecessários
            await page.route(
                "**/*",
                lambda route, request: route.abort() 
                if request.resource_type in ["image", "media", "font"] 
                else route.continue_()
            )

            await page.goto(url, wait_until="domcontentloaded", timeout=20000)

            # Espera o container principal
            await page.wait_for_selector("#TextoDigitadoTxt", state="attached", timeout=10000)
            
            # TRUQUE: Clicar no elemento para garantir que o JS do site ative
            try:
                await page.click("#TextoDigitadoTxt", timeout=2000)
            except:
                pass

            # Loop de verificação (Polling)
            conteudo = ""
            titulo = ""
            
            # Tenta ler o valor por até 5 segundos
            for _ in range(10):
                dados = await page.evaluate("""() => {
                    const textoEl = document.getElementById("TextoDigitadoTxt");
                    const tituloEl = document.getElementById("PastaTxt");
                    return {
                        conteudo: textoEl ? textoEl.value.trim() : "",
                        titulo: tituloEl ? tituloEl.innerText.trim() : ""
                    }
                }""")
                
                conteudo = dados['conteudo']
                titulo = dados['titulo']
                
                # Se achou conteudo relevante, para de esperar
                if conteudo and len(conteudo) > 15: 
                    break
                
                await asyncio.sleep(0.5)

            # VALIDAÇÃO RÍGIDA:
            # Se o conteúdo for vazio ou muito pequeno, consideramos falha
            if not conteudo or len(conteudo) < 5:
                raise ValueError(f"Conteúdo vazio ou muito curto (len={len(conteudo)})")

            return {
                "pagina": pagina,
                "titulo": titulo,
                "conteudo": conteudo,
                "status": "sucesso"
            }

        except Exception as e:
            # Só imprime erro se for a última tentativa
            if tentativa == max_retries - 1:
                print(f"❌ Falha final pág {pagina}: {str(e)[:50]}")
                return {
                    "pagina": pagina,
                    "titulo": titulo if 'titulo' in locals() else None, # Tenta salvar o título pelo menos
                    "conteudo": None,
                    "status": f"erro: {str(e)[:50]}"
                }
            
            # Backoff exponencial: espera mais a cada erro (2s, 4s, 6s...)
            wait_time = (tentativa + 1) * 2 + random.uniform(0, 1)
            await asyncio.sleep(wait_time)
            
        finally:
            await page.close()

async def main():
    resultados = []

    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        # User agent ajuda a evitar bloqueios ou comportamentos mobile
        context = await browser.new_context(
            user_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/120.0.0.0 Safari/537.36"
        )
        
        sem = asyncio.Semaphore(PARALLEL_PAGES)

        async def worker(pagina):
            async with sem:
                return await extract_page(context, pagina)

        tasks = [worker(i) for i in range(1, TOTAL_PAGINAS + 1)]

        for coro in tqdm(asyncio.as_completed(tasks), total=len(tasks), desc="Extraindo"):
            r = await coro
            resultados.append(r)

        await browser.close()

    df = pd.DataFrame(resultados).sort_values("pagina")
    
    # Salvar
    df.to_csv("dados_docvirt_v3.csv", index=False, encoding="utf-8-sig")

    # Diagnóstico
    vazios = df[df['conteudo'].isna() | (df['conteudo'] == "")]
    if not vazios.empty:
        print(f"⚠️ Atenção: {len(vazios)} páginas ainda estão sem conteúdo:")
        print(vazios['pagina'].tolist())
    else:
        print("✅ Todas as páginas foram extraídas com conteúdo!")

    return df

# await main()
```
